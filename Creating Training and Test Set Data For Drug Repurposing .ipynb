According to Martin et al 2009 one of the classifications of medications that is used to treat anxiety is called monoamines. It was discussed that anxiety medications that were affected targeted and inhibited the neurotransmitter for serotinin, known as 5HT. Based on this we propose that other drugs that inhibit 5HT could be good candidates for use as anxiety medications. 

Using ChEMBL 5HT was searched and then the recorded activities were investigated. Based on Ligand-Efficiency for Drug Discovery by Abad-Zapatero, some of the key components to consider for drug discovery are Ki, EC50, and IC50. 

Some important things to note are there are two types of assays that are being performed to collect this data: Binding and Functional. As defined on the ChEMBL Interface Documentation page for FAQs: "Binding (B) - Data measuring binding of compound to a molecular target, e.g. Ki, IC50, Kd and Functional (F) - Data measuring the biological effect of a compound, e.g. %cell death in a cell line, rat weight." (https://chembl.gitbook.io/chembl-interface-documentation/frequently-asked-questions/chembl-data-questions) 
Although we plan on focusing on Binding assays, the data has not yet been sorted according to this. 

Current plan: Label molecules as active or inactive based on optimum characteristics (ie a Ki below a certain value). Our goal is to find drugs that we think would be shown to bind to 5HT if a binding assay were to be performed on this assay. If a drug binds to and inhibits 5HT it could be pursued as a possible anxiety medication in the future.

IC50, which stands for "Inhibitory Concentration 50%," is a measure of the concentration of the inhibitor necessary to reduce the biological activity to 50% of the normal, unihibited level. In drug discovery this can be thought of a measure of efficacy. 

```python
df5HT_IC50_ChEMBL = pd.read_csv('IC50_HT.csv', sep = ';')
df5HT_IC50_ChEMBL.drop(['Molecule Name','Assay Description','Potential Duplicate','Molecule Max Phase','Comment','Assay Organism','Document ChEMBL ID','Target Type','Source ID','Document Journal','BAO Format ID','Target ChEMBL ID','Target Name','Target Organism','Standard Relation','AlogP','Data Validity Comment','Uo Units','BAO Label','Assay Tissue ChEMBL ID','Assay Tissue Name','Assay Cell Type','Assay Subcellular Fraction', '#RO5 Violations','Source Description','Document Year','Cell ChEMBL ID','Properties','Assay Parameters','Assay Variant Accession','Assay Variant Mutation'], axis=1, inplace=True)
df5HT_IC50_ChEMBL
```

EC50, which stands for "Effective Concentration 50%," is a measure of the concentration of a drug required to achieve 50% of the maximum effect of the drug. This makes EC50 a good measurement of the potency of a drug. However, because EC50 could apply to stimulating or inhibiting a receptor it is essential to make this distinction when creating our dataframes. 

```python
df5HT_EC50_ChEMBL = pd.read_csv('EC50_HT.csv', sep = ';')
df5HT_EC50_ChEMBL.drop(['Ligand Efficiency BEI','Ligand Efficiency LE','Ligand Efficiency LLE','Ligand Efficiency SEI','Comment','Assay Organism','Document ChEMBL ID','Target Type','Source ID','Document Journal','BAO Format ID','Target ChEMBL ID','Target Name','Target Organism','Standard Relation','AlogP','Data Validity Comment','Uo Units','BAO Label','Assay Tissue ChEMBL ID','Assay Tissue Name','Assay Cell Type','Assay Subcellular Fraction', '#RO5 Violations','Source Description','Document Year','Cell ChEMBL ID','Properties','Assay Parameters','Assay Variant Accession','Assay Variant Mutation'], axis=1, inplace=True)
df5HT_EC50_ChEMBL
```

```python
EC_Inhibition = df5HT_EC50_ChEMBL[df5HT_EC50_ChEMBL['Assay Description'].str.contains('Inhibition',na=False)]
EC_Inhibition
```
